```python
#to check the missing values; so creating a new variable and adding logic
features_na=[features for features in data.columns if data[features].isnull().sum()>1]


for feature in features_na:
   print(feature, np.round(data[feature].isnull().mean(),4), '%missing values')
```

```python
for feature in features_na:
   data1 = data.copy()
# if feature is null/missing give 1 else 0
data1[feature]=np.where(data[feature].isnull(), 1, 0)
#plot bar graph, grouping by feature with SalePrice
data1.groupby(feature)['SalePrice'].median().plot.bar()
plt.title(feature)
plt.show()
```

```python
#to find numeriacal feature in the dataset where the datatype is not equal to object
num_fea=[feature for feature in data.columns if data[feature].dtypes != 'O']
print('Num of numerical variables:' , len(num_fea))
data[num_fea].head()
```

```python
#to find only the year columns in the dataset 
year_fea=[feature for feature in num_fea if 'Yr' in feature or 'Year' in feature]
year_fea
```

```python
for feature in year_fea:
    print(feature, data[feature].unique())
```

```python
data.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title("House Price vs YrSold")
```

```python
#to find the difference between the YrSold and the year features
for feature in year_fea:
    if feature != 'YrSold':
        data1=data.copy()
        data[feature]=data['YrSold']-data[feature]
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
#Numeriacal variables is of 2 types i.e continuous and discreet
#For discrete feature, considering the num_fea the length of feaure must be less tha 25 and not in year+Id
dis_fea=[feature for feature in num_fea if len(data1[feature].unique())<25 and feature not in year_fea+['Id']]
print("Discrete variables count: {}".format(len(dis_fea)))
```

```python
dis_fea
```

```python
#Relation b/n dis_fea and SalePrice
for feature in dis_fea:
    data1=data.copy()
    data1.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.show()
```

```python
#creating continuous feature, where feature is not in dis,year+Id
cont_fea=[feature for feature in num_fea if feature not in dis_fea+year_fea+['Id']]
print("Continuous Feature {}".format(len(cont_fea)))
```

```python
#Plotting a histogram
for feature in cont_fea:
    data1=data.copy()
    data1[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.show()
```

```python
#Lograthmic Transformation

for feature in cont_fea:
    data1=data.copy()
    if 0 in data1[feature].unique():
        pass
    else:
        data1[feature]=np.log(data1[feature])
        data1['SalePrice']=np.log(data1['SalePrice'])
        plt.scatter(data1[feature],data1['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
########outliers
for feature in cont_fea:
    data1=data.copy()
    if 0 in data1[feature].unique():
        pass
    else:
        data1[feature]=np.log(data1[feature])
        data1.boxplot(column=feature)
        plt.ylabel(feature)
        plt.show()
```

```python
#####Categorial Variables
cat_fea=[feature for feature in data.columns if data1[feature].dtypes== 'O']
cat_fea
```

```python
for feature in cat_fea:
    print('The feature is {} and the no. of categories are{}'.format(feature,len(data[feature].unique())))
```
