```python
###Feature  Engineering
import pandas as pd
import numpy as  np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv(r'D:\dataScience\Advanced-House-Price-Prediction-\Advanced-House-Price-Prediction-\train.csv')
dataset.head(5)
```

```python
#Split the data into train and test to prevent data leakage
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(dataset,dataset['SalePrice'], test_size=0.1,random_state=0)
```

```python
X_train.shape, X_test.shape
```

```python
#find missing values
fea_nan=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1 and dataset[feature].dtype=='O']
for feature in fea_nan:
    print("{}: {}% missing values".format(feature,np.round(dataset[feature].isnull().mean(),4)))
```

```python
#Replace Missing Value with fill
def replace_cat_fea(dataset, fea_nan):
    data=dataset.copy()
    data[fea_nan]=data[fea_nan].fillna('Missing')
    return data
dataset=replace_cat_fea(dataset,fea_nan)
dataset[fea_nan].isnull().sum()
```

```python
num_fea=['LotFrontage', 'LotArea']
for feature in num_fea:
    dataset[feature]=np.log(dataset[feature])
dataset.head(5)
```

```python
cat_fea=[feature for feature in dataset.columns if dataset[feature].dtype=='O']
cat_fea
```

```python
#get percentage of each category considering the dataset
for feature in cat_fea:
    #Groping by feature and taking the saleprice no. of records and dividing it by the overall dataset length
    temp=dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    #if the percentage is > 0.01
    temp_df=temp[temp>0.01].index
    #if the dataset[feature] isin temp_df then take that dataset[feature] else convert it into Rere_var/any name
    dataset[feature]=np.where(dataset[feature].isin(temp_df),dataset[feature],'Rare_var')
    
```

```python
dataset.head(50)



```

```python
#Feature Scaling
fea_scale=[feature for feature in dataset.columns if feature not in ['Id', 'SalePrice']]

from sklearn.preprocessing import MinMaxScaler

#creating an object

scaler=MinMaxScaler()
scaler.fit(dataset[fea_scale])
```

```python
fea_scale
```

```python
#transforming the train and test dataset and adding the Id and SalePrice variables
data = pd.concat([dataset[['Id', 'SalePrice']].reset_index(drop=True), pd.DataFrame(scaler.transform(dataset[fea_scale]), 
                                                   columns=fea_scale)],axis=1)

data
```
